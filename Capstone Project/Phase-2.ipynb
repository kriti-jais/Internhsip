```python
# GROUPING THE DATA YEARWISE FOR EACH STATE AND TAKING THE SUM(), <BR>
```

```python
uni=human_right['STATE/UT'].unique().tolist()
ye=human_right['Year'].unique().tolist()
```

```python
state=[]
year=[]
cases=[]
police_charge=[]
police_convict=[]
for i in uni:
    for j in ye:
        temp=human_right[(human_right['STATE/UT']==i)&(human_right['Year']==j)].groupby(by='Year').sum().reset_index()
        if len(temp)==1:
            state.append(i)
            year.append(temp['Year'][0])
            cases.append(temp['Cases_Registered_under_Human_Rights_Violations'][0])
            police_charge.append(temp['Policemen_Chargesheeted'][0])
            police_convict.append(temp['Policemen_Convicted'][0])
        else:
            state.append(i)
            year.append(j)
            cases.append(0)
            police_charge.append(0)
            police_convict.append(0)
```

```python
mod_human=pd.DataFrame({"STATE/UT":state,"Year":year,"Cases_Registered_under_Human_Rights_Violations":cases,"Policemen_Chargesheeted":police_charge,"Policemen_Convicted":police_convict})
```

```python
def ff(data):
    name=[x for x in globals() if globals()[x] is data][0]
    unwanted=['STATE/UT','Year']
    col=data.columns.tolist()
    col=[i for i in col if i not in unwanted]
    new_col={i:i+'_'+str(name).upper() for i in col}
    data.rename(columns=new_col,inplace=True)
```

# 2.5) Analysis of population vs overall crime!!

<B> CONSIDERING OVERALL POPULATION!!<BR>

```python
plt.figure(figsize=(15,8),facecolor='Yellow')
sns.scatterplot(x=mer_data['Population Total_DIST'],y=mer_data['OVERALL CRIME'])
plt.xlabel("TOTAL POPULATION")
plt.ylabel("OVERALL CRIEMES!!")
plt.title("OVERALL CRIMES VS THE TOTAL POPULATION")
plt.grid()
plt.show()
```

The relationship between population growth and crime rates is a complex issue with many factors that contribute to it. While there is some evidence to suggest that there may be a correlation between population growth and crime rates, it is not a direct relationship.<BR><BR>
One reason why an increase in population could lead to an increase in crime is the concept of strain theory. This theory suggests that when people experience strains such as poverty, unemployment, or lack of resources, they may become frustrated and turn to criminal behavior as a means of coping. When a population grows, there may be more competition for resources and opportunities, leading to more strains and potentially more criminal behavior.<BR><BR>
Additionally, as the population increases, so does the diversity of the community. Diversity can lead to cultural clashes and conflicts, which may contribute to an increase in crime. Moreover, a larger population can lead to a breakdown in social networks and community ties, making it easier for criminals to go unnoticed and commit crimes.<BR><BR>
