# Q.1) Python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.


```python
driver.get("https://www.naukri.com")
```

```python
search=driver.find_element(By.XPATH,'/html/body/div[1]/div[6]/div/div/div[6]')
search.click()
```

```python
job_title=[]
for i in driver.find_elements(By.XPATH,'//a[@class="title ellipsis"]'):
    job_title.append(i.text)
    
job_title[0:9]
```

```python
job_location=[]
for i in driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft locWdth"]'):
    job_location.append(i.text)
    
job_location[0:9]    
```

```python
experience=[]
for i in driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft expwdth"]'):
    experience.append(i.text)
    
experience[0:9]
```

```python
company=[]
for i in driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]'):
    company.append(i.text)
    
company[0:9]
```

```python
df1=pd.DataFrame({"Title":job_title,"Location":job_location,"Company":company,"Experience":experience})
df1[0:10]
```

```python
driver.get("https://www.azquotes.com")
```

```python
top_quote=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a')
top_quote.click()
```

```python
Quote=[]
author=[]
typeofquote=[]
start=0
end=10

for page in range(start,end):
    pr=driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in pr:
        Quote.append(i.text)     
      
    aut=driver.find_elements(By.XPATH,'//div[@class="author"]')
    for i in aut:
        author.append(i.text)     
    

    toq=driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in toq:
        typeofquote.append(i.text)     
    nextbutton=driver.find_element(By.XPATH,'//li[@class="next"]')
    nextbutton.click ()
    time.sleep(2)
```

```python
print(len(Quote),len(author),len(typeofquote))
```
